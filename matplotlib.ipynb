```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y=x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x, y, 'r') #'r' adalah warna merah
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('String Title Here')
plt.show()
```

```python
# plt.sublot(rows, ncols, plot_number)
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r--')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'g*-');
```

```python
#create figure (empty canvas)
fig = plt.figure()

# Add set of axe to figure
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #left, bottom, width, height (range 0 to 1)

# Plot on that set of axes
axes.plot(x, y, 'b')
axes.set_xlabel('Set X label')
axes.set_ylabel('Set y label')
axes.set_title('Set Title')
```

```python
#create figure (empty canvas)
fig = plt.figure()

# Add set of axe to figure
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # inset axes

# Larger Figure Axes 1
axes1.plot(x, y, 'b')
axes1.set_xlabel('X_Label_axes2')
axes1.set_ylabel('Y_Label_axes2')
axes1.set_title('Axes 2 Title')

# Insert Figure Axes 2
axes2.plot(y, x, 'r')
axes2.set_xlabel('X_Label_axes2')
axes2.set_ylabel('Y_Label_axes2')
axes2.set_title('Axes 2 Title')
```

```python
# Use similiar to plt.figure() except use tuple unpacking to grab fig and axes
fig, axes = plt.subplots()

# Now use the axes object to add stuff to plot
axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```

```python
# Empty canvas of 1 by 2 subplots
fig, axes = plt.subplots(nrows=2, ncols=1)
```

```python
axes
```

```python
for ax in axes:
    ax.plot(x, y, 'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')
    
# Display the figure object
fig
```

```python
# Empty canvas of 1 by 2 subplots
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(12,5), dpi=100)

ax0 = axes[0][0]
ax0.plot(x, y, 'r')
ax0.set_xlabel('x')
ax0.set_ylabel('y')
ax0.set_title('title')

ax1 = axes[0][1]
ax1.plot(y, y, 'b')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('title')

ax2 = axes[0][2]
ax2.plot(x, y, color='red', label='expense', lw=7)
ax2.set_xlabel('holidays')
ax2.set_ylabel('expenses')
ax2.set_title('holiday vs expense')
ax2.legend()

ax3 = axes[1][0]
ax3.plot(x, x**3, 'g', label='x**3')
ax3.plot(x, y, 'r--', label='x**2')
ax3.set_xlabel('x')
ax3.set_ylabel('y')
ax3.set_title('title')
ax3.legend()

from random import sample
data = sample(range(1, 1000), 100)
ax3 = axes[1][1]
ax3.hist(data)

ax4 = axes[1][2]
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.boxplot(data,vert=True, patch_artist=True)

plt.tight_layout()
```
